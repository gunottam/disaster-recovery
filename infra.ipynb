```python
import os
import zipfile
import json
import joblib
import pandas as pd
import geopandas as gpd
import seaborn as sns
import folium
import matplotlib.pyplot as plt
import rasterio
from matplotlib.lines import Line2D
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score










import osmnx as ox

# --- OSM Data Download ---
# Buildings
buildings = ox.features_from_place('Kathmandu, Nepal', tags={'building': True})
buildings.to_file('kathmandu_buildings.geojson', driver='GeoJSON')
print("✅ Buildings GeoJSON saved.")

# Hospitals
hospitals = ox.features_from_place('Kathmandu, Nepal', tags={'amenity': 'hospital'})
hospitals.to_file('kathmandu_hospitals.geojson', driver='GeoJSON')
print("✅ Hospitals GeoJSON saved.")

# --- Load and Plot OSM Data ---
buildings_gdf = gpd.read_file('kathmandu_buildings.geojson')
hospitals_gdf = gpd.read_file('kathmandu_hospitals.geojson')

fig, ax = plt.subplots(figsize=(12, 12))
buildings_gdf.plot(ax=ax, color='lightgray', linewidth=0.2)
hospitals_gdf.plot(ax=ax, color='red', markersize=10)
legend_elements = [
    Line2D([0], [0], marker='o', color='w', markerfacecolor='red', markersize=10, label='Hospitals'),
    Line2D([0], [0], color='lightgray', lw=4, label='Buildings')
]
plt.title('Hospitals and Buildings in Kathmandu', fontsize=16)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(handles=legend_elements, loc='upper right')
plt.grid(True)
plt.show()

# --- NASA Damage Proxy Map ---
file_path = 'nasa_damage.tif'
try:
    damage_map = rasterio.open(file_path)
    damage_data = damage_map.read(1)
    plt.imshow(damage_data, cmap='hot')
    plt.title("NASA ARIA Damage Proxy Map")
    plt.colorbar()
    plt.show()
except FileNotFoundError:
    print(f"Error: The file {file_path} was not found.")

# --- Extract and Load CSV ---
zip_path = r'C:/Users/anany/Downloads/eq2015.csv.zip'
extract_folder = r'C:/Users/anany/Downloads/eq2015_extracted'
os.makedirs(extract_folder, exist_ok=True)
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_folder)
print("Extracted files to:", extract_folder)

# --- Load and Clean CSV ---
csv_file_path = os.path.join(extract_folder, 'eq2015.csv')
df = pd.read_csv(csv_file_path)
df['damage_grade'] = df['damage_grade'].fillna('Unknown')
df['technical_solution_proposed'] = df['technical_solution_proposed'].fillna('Unknown')
df['plan_configuration'] = df['plan_configuration'].fillna('Unknown')

# --- Filter & Analyze ---
df_filtered = df[['building_id', 'damage_grade', 'district_id', 'vdcmun_id', 'condition_post_eq']]
damage_count = df_filtered['damage_grade'].value_counts()
district_damage = df_filtered.groupby('district_id')['damage_grade'].value_counts().unstack(fill_value=0)

plt.figure(figsize=(10, 6))
damage_count.plot(kind='bar', color='skyblue')
plt.title('Damage Distribution by Grade')
plt.xlabel('Damage Grade')
plt.ylabel('Number of Buildings')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 8))
sns.heatmap(district_damage, annot=True, cmap="YlGnBu", fmt="d")
plt.title('Damage by District and Grade')
plt.xlabel('Damage Grade')
plt.ylabel('District ID')
plt.show()

# Save filtered data
filtered_csv_path = 'cleaned_eq2015.csv'
df_filtered.to_csv(filtered_csv_path, index=False)

# --- Machine Learning Model ---
categorical_columns = ['land_surface_condition', 'foundation_type', 'roof_type', 'ground_floor_type',
                       'other_floor_type', 'position', 'plan_configuration', 'condition_post_eq', 'superstructure']
for col in categorical_columns:
    df[col] = df[col].astype(str).fillna('Unknown')
    df[col] = LabelEncoder().fit_transform(df[col])

le_damage = LabelEncoder()
df['damage_grade'] = le_damage.fit_transform(df['damage_grade'])

features = ['count_floors_pre_eq', 'count_floors_post_eq', 'age_building', 'plinth_area_sq_ft',
            'height_ft_pre_eq', 'height_ft_post_eq'] + categorical_columns
X = df[features]
y = df['damage_grade']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print(f"Model Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=le_damage.classes_))

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=le_damage.classes_, yticklabels=le_damage.classes_)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()

# Feature Importance
importance_df = pd.DataFrame({
    'Feature': features,
    'Importance': model.feature_importances_
}).sort_values(by='Importance', ascending=False).head(10)
sns.barplot(x='Importance', y='Feature', data=importance_df, palette='viridis')
plt.title("Top 10 Important Features")
plt.tight_layout()
plt.show()

# Save model and predictions
joblib.dump(model, 'earthquake_damage_model.pkl')
df['predicted_damage_grade'] = le_damage.inverse_transform(model.predict(X))
df.to_csv('predictions.csv', index=False)
print("Predictions saved to 'predictions.csv'.")

# --- UNOSAT Damage Shapefile ---
zip_file_path = 'C:/Users/anany/Downloads/all_damage_nepal.zip'
extraction_folder = 'C:/Users/anany/Downloads/extracted/'
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extraction_folder)

shp_file = None
for root, dirs, files in os.walk(extraction_folder):
    for file in files:
        if file.endswith('.shp'):
            shp_file = os.path.join(root, file)
            break

if shp_file:
    gdf = gpd.read_file(shp_file)
    if gdf.crs != "EPSG:4326":
        gdf = gdf.to_crs("EPSG:4326")
    geojson_data = json.loads(gdf.to_json())
    m = folium.Map(location=[27.7, 85.3], zoom_start=8, tiles='CartoDB positron')
    folium.GeoJson(
        geojson_data,
        name="UNOSAT Damage",
        tooltip=folium.GeoJsonTooltip(
            fields=["Damage", "Damage_Int"],
            aliases=["Damage Level", "Damage Intensity"],
            sticky=True
        ),
        style_function=lambda feature: {
            'fillColor': '#ff0000' if feature["properties"]["Damage"] == "AFFECTED" else '#gray',
            'color': 'red',
            'weight': 0.5,
            'fillOpacity': 0.5
        }
    ).add_to(m)
    folium.LayerControl().add_to(m)
    m.save("unosat_damage_map.html")
    print("UNOSAT damage map saved to 'unosat_damage_map.html'.")
else:
    print("No .shp file found in the extracted contents.")
```
